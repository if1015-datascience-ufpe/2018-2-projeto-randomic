## Extração de conteúdo da página

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
pages = [(bairro, requests.get(url)) for bairro, url in bairros.items()]
```

```python
ds = []
for bairro, page in pages:
    if page.status_code == 200:
        soup = BeautifulSoup(page.content, 'html.parser')
        data = soup.find('div', class_='content-text text-servico')
        
        ps = [p for p in data.find_all('p') if len(p.text) > 1]
        
        p = str(ps[0].text)
        RPA = int(p[p.find('RPA') + 4:p.find(', Micro')])
        microrregiao = float(p[p.find('Microrregião: ') + 14:][:3])
        dist = p.replace('(km):', '(km)1:')
        dist = dist[dist.find('(km)1') + 7: dist.find('Área Territorial')]
        dist = float(dist.replace(' ', '').replace(',', '.'))
        hect = float(p[p.find('(hectare)') + 12: p.find('População')].replace('.', '').replace(',', '.').replace(' ', ''))
        hab = int(p[p.rfind(':') + 2:].replace('.', '').replace('habitantes', '').replace(' ', ''))
        
        tables = data.find_all('table')
        
        genre_data = tables[0].find_all('td')
        mN = int(genre_data[4].text.replace('.', ''))
        mP = float(genre_data[5].text.replace(',', '.'))
        fN = int(genre_data[7].text.replace('.', ''))
        fP = float(genre_data[8].text.replace(',', '.'))

        habs = []
        hab_data = tables[1].find_all('td')
        for i, j in [(4, 5), (7, 8), (10, 11), (13, 14), (16, 17), (19, 20)]:
            habs.append(int(hab_data[i].text.replace('.', '')))
            habs.append(float(hab_data[j].text.replace(',', '.')))

        ethnics = []
        ethnic_data = tables[2].find_all('td')
        for i in range(3, 12, 2):
            ethnics.append(float(ethnic_data[i].text.replace('.', '').replace(',', '.')))
        
        ps = [p.text for p in ps]
        
        if len(ps) < 8:
            txt = "\n".join(ps)
            txt = txt.replace('Taxa Média Geométrica', '\nTaxa Média Geométrica')
            txt = txt.replace('Taxa Media Geométrica', '\nTaxa Media Geométrica')
            txt = txt.replace('Densidade Demográfica', '\nDensidade Demográfica')
            txt = txt.replace('%Densidade', '\n%Densidade')
            txt = txt.replace('Densidade demográfica', '\nDensidade demográfica')
            txt = txt.replace('Domicílios (n', '\nDomicílios (n')
            txt = txt.replace('Domicílios(n', '\nDomicílios(n')
            ps = [d for d in txt.split('\n') if len(d) > 1]
        
        try:
            value = float(ps[1].replace(',', '.'))
            del ps[1]
        except Exception:
            pass
        if ps[1].find('População Residente') >= 0:
            hab = int(ps[1][ps[1].rfind(':') + 2:].replace('.', '').replace('habitantes', '').replace(' ', ''))
            del ps[1]
        
        try:
            alfa = float(ps[1][ps[1].rfind(':') + 1:].replace(',', '.')[:-1])
        except Exception:
            alfa = float(ps[1][ps[1].rfind('(%)4') + 4:].replace(',', '.'))
            
        mean = float(ps[2][ps[2].rfind(':') + 1:].replace(',', '.').replace('(%)', '').replace('%', '').replace(' ', ''))
        dens = float(ps[3][ps[3].rfind(':') + 1:].replace('.', '').replace(',', '.').replace(' %', ''))
        try:
            domc = int(ps[4][ps[4].rfind(':') + 1:].replace('.', '').replace(' ', ''))
            media = None
            propM = None
            rend = None
            
        except Exception:
            txt = "\n".join(ps)
            txt = txt.replace('Média de Moradores', '\nMédia de Moradores')
            txt = txt.replace('Média de moradores', '\nMédia de moradores')
            txt = txt.replace('Proporção de Mulheres', '\nProporção de Mulheres')
            txt = txt.replace('Valor do Rendimento Nominal', '\nValor do Rendimento Nominal')
            txt = txt.replace('Zonas Especiais', '\nZonas Especiais')
            ps = [d for d in txt.split('\n') if len(d) > 1]
            
            domc = int(ps[4][ps[4].rfind(':') + 1:].replace('.', '').replace(' ', ''))
            media = float(ps[5][ps[5].rfind(':') + 1:].replace(',', '.'))
            propM = float(ps[6][ps[6].rfind(':') + 1:].replace(',', '.'))
            rend = float(ps[7][ps[7].rfind('R$') + 2:].replace('.', '').replace(',', '.'))
        
        items = data.find_all('li')
        if len(items) == 3:
            media = float(items[0].text[items[0].text.rfind(':') + 1:].replace(',', '.'))
            try:
                propM = float(items[1].text[items[1].text.rfind(':') + 1:].replace(',', '.').replace('%', '').replace(' ', ''))
            except Exception:
                propM = float(items[1].text[items[1].text.rfind('(%)') + 3:].replace(',', '.').replace('%', '').replace(' ', ''))
            rend = float(items[2].text[items[2].text.rfind('R$') + 3:].replace('.', '').replace(',', '.'))
            
        ds.append([bairro, RPA, microrregiao, dist, hect, hab, mN, mP, fN, fP] + habs + ethnics + [alfa, mean, dens, domc, media, propM, rend])
```

```python
columns = [
    'BAIRRO', 
    'RPA', 
    'MICROREGIAO', 
    'DIST_MARCO_ZERO', 
    'HECTARES', 
    'HABITANTES', 
    'NUM_HOMENS', 
    'POR_HOMENS',
    'NUM_MULHERES',
    'POR_MULHERES',
    '0_4ANOS',
    '0_4ANOS%',
    '5_14ANOS',
    '5_14ANOS%',
    '15_17ANOS',
    '15_17ANOS%',
    '18_24ANOS',
    '18_24ANOS%',
    '25_59ANOS',
    '25_59ANOS%',
    '60_ANOS',
    '60_ANOS%',
    'BRANCA',
    'PRETA',
    'PARDA',
    'AMARELA',
    'INDIGENA',
    'ALFABETIZADOS', 
    'CRESCIMENTO_ANUAL', 
    'DENSIDADE', 
    'DOMICILIOS', 
    'MEDIA_MORADORES', 
    'MULHERES_RESPONSAVEIS', 
    'RENDA']
```

```python
df = pd.DataFrame(ds, columns=columns)
```
